```python
import matplotlib.pyplot as plt
from skimage.color import rgb2gray
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
import matplotlib.image as maing
img = cv.imread('q1.jpeg', 0)
# cv.imshow('code',img)

ret, th1 = cv.threshold(img, 0, 255, cv.THRESH_BINARY+cv.THRESH_OTSU)
adap = cv.adaptiveThreshold(img, 255, cv.ADAPTIVE_THRESH_GAUSSIAN_C,
                            cv.THRESH_BINARY, 31, 5)
cv.imshow("", th1)
cv.waitKey(0)
cv.imshow("", adap)
cv.waitKey(0)
cv.destroyAllWindows()
plt.imshow(img, "gray")
```
