```python
import matplotlib.pyplot as plt
from skimage.color import rgb2gray
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
import matplotlib.image as maing
img = cv.imread('q1.jpeg', 0)
# cv.imshow('code',img)

# ret, th1 = cv.threshold(img, 0, 255, cv.THRESH_BINARY+cv.THRESH_OTSU)
adap = cv.adaptiveThreshold(img, 255, cv.ADAPTIVE_THRESH_GAUSSIAN_C,
                            cv.THRESH_BINARY_INV, 31, 5)
ret, th1 = cv.threshold(img, 127, 255, cv.THRESH_BINARY+cv.THRESH_OTSU)
th2 = cv.adaptiveThreshold(img, 255, cv.ADAPTIVE_THRESH_MEAN_C,
                           cv.THRESH_BINARY, 11, 2)
tiltes = ["adap","simple","adapmean"]
phots = [adap,th1,th2]
for i in range(3):
    plt.subplot(2, 2, i+1), plt.imshow(phots[i], 'gray')
    plt.title(tiltes[i])
plt.show()

```
