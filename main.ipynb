```python
import matplotlib.pyplot as plt
from skimage.color import rgb2gray
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
import matplotlib.image as maing
img = cv.imread('ac.jpeg', 0)
# cv.imshow('code',img)

# ret, th1 = cv.threshold(img, 0, 255, cv.THRESH_BINARY+cv.THRESH_OTSU)
adap = cv.adaptiveThreshold(img, 255, cv.ADAPTIVE_THRESH_GAUSSIAN_C,
                            cv.THRESH_BINARY_INV, 31, 5)
ret, th1 = cv.threshold(img, 127, 255, cv.THRESH_BINARY+cv.THRESH_OTSU)
th2 = cv.adaptiveThreshold(img, 255, cv.ADAPTIVE_THRESH_MEAN_C,
                           cv.THRESH_BINARY, 11, 2)
tiltes = ["adap","simple","adapmean"]
phots = [adap,th1,th2]
for i in range(3):
    plt.subplot(2, 2, i+1), plt.imshow(phots[i], 'gray')
    plt.title(tiltes[i])
plt.show()

```

```python
from PIL import ImageEnhance
from PIL import Image
image = Image.open("newone.jpg")
image.show()

# enhance
ImageEnhance.Contrast(image)
curr_con = ImageEnhance.Contrast(image)
new_con = 1.0
# Contrast enhanced by a factor of 0.3
img_contrasted = curr_con.enhance(new_con)
# shows updated image in image viewer
# img_contrasted.show()

# sharpness
curr_sharp = ImageEnhance.Sharpness(img_contrasted)
new_sharp = 2

# Sharpness enhanced by a factor of 8.3
img_sharped = curr_sharp.enhance(new_sharp)

# shows updated image in image viewer
# img_sharped.show()

# brightness
curr_bri = ImageEnhance.Brightness(img_sharped)
new_bri = 3

# Brightness enhanced by a factor of 2.5
img_brightened = curr_bri.enhance(new_bri)

# shows updated image in image viewer
# img_brightened.show()

curr_col = ImageEnhance.Color(img_brightened)
new_col = 9.5

# Color level enhanced by a factor of 2.5
img_colored = curr_col.enhance(new_col)

# shows updated image in image viewer
img_colored.show()
```
